```python
# Start coding here
# Use as many cells as you need
def validate_user(name,email,password):
    if not validate_name(name):
        raise ValueError('The name chould be a string with at least 3 caracters')
    elif not validate_email(email):
        raise ValueError('The email is not valid')
    elif not validate_password(password):
        raise ValueError('The password chould contains at least an upper caracter and a number')
    return True
        
```

```python
def register_user(name,email,password):
    if validate_user(name,email,password):
        user_dict={}
        user_dict['name']=name
        user_dict['email']=email
        user_dict['password']=password
        return user_dict
    else:
        return False
```
