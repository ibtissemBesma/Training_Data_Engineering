Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

```python
best_math_schools=schools[['school_name','average_math']][schools['average_math']>(0.8*800)].sort_values('average_math',ascending=False)
best_math_schools
```

```python
top_10_schools=pd.DataFrame()
top_10_schools['school_name']=schools['school_name']
top_10_schools['total_SAT']=schools['average_math']+schools['average_reading']+schools['average_writing'] 
```

```python
top_10_schools=top_10_schools.sort_values('total_SAT',ascending=False)
top_10_schools=top_10_schools.nlargest(10,'total_SAT')
top_10_schools
```

```python
schools['total_SAT']=schools['average_math']+schools['average_reading']+schools['average_writing'] 
```

```python
temp_df=pd.DataFrame()
temp_df=schools.groupby('borough')['total_SAT'].agg(['std','mean']).sort_values(['std','mean'],ascending=[False,False])
temp_df
```

```python
temp_df=pd.DataFrame(temp_df)
borough=temp_df.index[0]
std_SAT=temp_df['std'][0]
average_SAT=temp_df['mean'][0]
```

```python
num_schools=schools[schools['borough']==borough].value_counts('school_name').sum()
num_schools
```

```python
largest_std_dev_data=[borough,num_schools,average_SAT.round(),std_SAT.round()]
largest_std_dev=pd.DataFrame([largest_std_dev_data])
largest_std_dev.columns=['borough','num_schools','average_SAT','std_SAT']
largest_std_dev
```
